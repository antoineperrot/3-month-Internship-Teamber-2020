Admettons que 1 unité de distance = 1000 km ici et que l'on roule en moyenne à 80 km/h soit 1.33 km/min.
Il s'agit maintenant de construire la matrice des durées de trajets entre chaque destination possible


## Construction de la Distance Matrix (en minutes) : 

Structure : 
- colonne/ligne 0 : rdv1
- colonne/ligne 1 : rdv2
- colonne/ligne j>1 : selection indice j-2


Dans un premier temps on construit la vraie matrice des distances physiques, que l'on converti en nombre de minutes moyen, et que l'on vient ensuite modifier légèrement pour faire correspondre à des autoroutes, bouchons, etc... càd pour que cela ne correspondent pas linéairement à la distance physique.

```python
from scipy.spatial import distance_matrix
coords = np.vstack((df_rdv[['X','Y']].values,sel[['X','Y']].values))
DM = distance_matrix(coords,coords) * 1000 / 1.3
for i in range(1,DM.shape[0]):
    for j in range(0,i):
        u = np.random.normal(1,0.2)
        DM[i,j] *= u
        DM[j,i] *= u
DM = DM.astype(int)
DM.shape
```

# Calcul du temps restant

On considère que le commercial est disponible de 8h30 à 17h30.

Pour l'instant, nous ne visitons que le rdv1 et le rdv2 à des dates précises.

```python
pts = [0,4,2,6,8,1] #lieu visités dans cet ordre
```

```python
#Calcul du temps de trajet :
def temps_trajet(trajet,DistanceMatrix):
    t_trajet = 0
    for i in range(len(trajet)-1):
        t_trajet  += DistanceMatrix[trajet[i],trajet[i+1]]

    return t_trajet

t_trajet = temps_trajet(pts,DM)
print(t_trajet,'minutes de route pour ce trajet.')
```

```python
#Calcul du temps restant :
from datetime import timedelta
def temps_restant(df_rdv, pts, DistanceMatrix,
                  hor_deb = '08:30:00', hor_fin ='17:30:00', minutes_par_visite = 30,show=False):
    rdv1 = df_rdv.loc[0]
    rdv2 = df_rdv.loc[1]
    hor_deb = '08:30:00'
    hor_fin = '17:30:00'
    current_day = rdv1['Date fin'].date()
    a = pd.to_datetime(str(current_day) + ' '+ hor_fin)
    t_restant = 0
    t_restant += max((a - rdv1['Date fin'])/np.timedelta64(1,'m'),0)
    current_day += timedelta(days=1)
    while current_day != rdv2['Date début'].date():
        t_restant += (pd.to_datetime(hor_fin) - pd.to_datetime(hor_deb))/np.timedelta64(1,'m')
        current_day += timedelta(days=1)

    a = pd.to_datetime(str(current_day) + ' '+ hor_deb)

    t_restant += max( (rdv2['Date début'] - a) / np.timedelta64(1,'m'), 0)

    t_trajet = temps_trajet(pts,DistanceMatrix)
    t_surplace = minutes_par_visite * (len(pts) - 2 )
    if show :
        print("Au total dispo :",t_restant / 60 ,"h")
        print("Temps sur la route :",t_trajet/ 60 ,"h")
        print("Temps sur place :",t_surplace/ 60 ,"h")
        print('Restant :', (t_restant - t_trajet - t_surplace)/ 60 ,"h")

    
    return t_restant - t_trajet - t_surplace
temps_restant(df_rdv, pts, DM)
```

## Calcul du trajet optimal (recuit simulé) :

```python
def trajet_optimal(pts, DistanceMatrix, ITERATIONS_PAR_LIEU = 350):
    TEMPERATURE_INITIALE = 1
    TEMPERATURE_MIN = 1e-2
    POURCENTAGE_AVANT_TMIN = .999
    N_iterations = ITERATIONS_PAR_LIEU * (len(pts) - 2)
    dm = DistanceMatrix/ DistanceMatrix.max()
    x = np.array(pts)
    xmin = x
    Proba = np.zeros(N_iterations)
    E = np.zeros(N_iterations)
    E[0] = temps_trajet(x, dm)
    Emin = np.copy(E)
    T = TEMPERATURE_INITIALE
    decay = 1 - np.exp(np.log(TEMPERATURE_MIN/TEMPERATURE_INITIALE)/POURCENTAGE_AVANT_TMIN/N_iterations)

    for i in range(1,N_iterations):
        T = T*(1-decay) if T> TEMPERATURE_MIN else TEMPERATURE_MIN
        y = np.copy(x)
        j1 = np.random.randint(1,len(x)-1)
        j2 = np.random.randint(1,len(x)-1)
        y[j1] = x[j2]
        y[j2] = x[j1]
        Ey = temps_trajet(y, dm)
        DELTA = Ey - E[i-1]
        Proba[i] = np.exp( - DELTA / T)

        if np.random.rand() < Proba[i]:
            x = np.copy(y)
            E[i] = Ey
        else :
            E[i] = E[i-1]

        if E[i] < Emin[i-1]:
            Emin[i] = E[i]
            xmin = x
        else :
            Emin[i] = Emin[i-1]
            
    '''
    plt.figure(figsize=(8,6))
    plt.plot(range(N_iterations),E)
    plt.plot(range(N_iterations),Emin);
    '''

    return xmin
```

```python
pts = [0,4,2,3,6,8,1]
trajet = trajet_optimal(pts,DM) 
```

```python
def plot_trajet(trajet,selection,df_rdv):
    plt.figure(figsize=(8,8))
    plt.scatter(df_rdv['X'],df_rdv['Y'],c='r',label='RDV fermes')
    plt.scatter(selection['X'],selection['Y'],c='b',label='Sélection 1')
    plt.legend();
    plt.title("Carte complète");
    selection.reset_index(inplace=True,drop=True)

    for index, row in sel.iterrows():
        plt.annotate(row["Nom"],(row['X'],row['Y']))

    plt.plot(coords[trajet,0],coords[trajet,1],c='magenta');
    plt.show();
    return 
plot_trajet(trajet,sel,df_rdv)
```

```python
def candidat(trajet,DistanceMatrix):
    mini = None
    for pt in trajet :
        tmp = DistanceMatrix[pt,:].astype('float')
        tmp[trajet] = np.inf
        if mini is None or mini > np.min(tmp) :
            argmin = np.argmin(tmp)
            mini = np.min(tmp)
    #print(f"Prochain candidat : index {argmin - 2 } de la sélection, {argmin} dans la DM à {int(mini)} minutes en voiture d'un des points du trajet.")
    return argmin 
```

Notre itinéraire initial est  0 -> 1 -> 2 -> 3 -> 4 (nos rendez-vous fermes). L'idée va être d'ajouter des points sur la route tant que l'on en a le temps. 
On va partir du principe que le commercial travaille de 08h30 à 17h30.

Ci-dessous une fonction qui calcule le temps disponible entre deux rdv fermes compte tenu des horaires de travail du commercial

```python
def temps_dispo(row, debut_journee = '08:30:00', fin_journee   = '17:30:00'):
    current_day = row['Date fin rdv Origine'].date()

    if current_day  == row['Date début rdv Destination'].date():
        t_dispo = row['Date début rdv Destination'] - row['Date fin rdv Origine']
    else :
        a = pd.to_datetime(str(current_day) + ' '+ fin_journee)
        t_dispo = a - row['Date fin rdv Origine'] if a - row['Date fin rdv Origine'] >= datetime.timedelta(0) else datetime.timedelta(0)
        current_day += datetime.timedelta(days=1)
        while current_day != row['Date début rdv Destination'].date() :
            t_dispo += pd.to_datetime(fin_journee) - pd.to_datetime(debut_journee)
            current_day += datetime.timedelta(days=1)
        b = pd.to_datetime(str(current_day) + ' '+ debut_journee)
        t_dispo += row['Date début rdv Destination'] - b if row['Date début rdv Destination'] - b > datetime.timedelta(0) else datetime.timedelta(0)
    return t_dispo
```

On constitue un dataframe qui va renseigner les intervalles entre les rdv fermes afin d'ajouter si possible des points de passages entre ces rdv fermes.

On considère maintenant que visiter un prospect prend 30 minutes, et visiter un client 2h.

```python
InterRDV = pd.DataFrame()
InterRDV['Origine'] = data.loc[RDV[:-1],'Nom'].values
InterRDV['Origine Id'] = RDV[:-1]
InterRDV['Destination'] = data.loc[RDV[1:],'Nom'].values
InterRDV['Destination Id'] = RDV[1:]
InterRDV['Date fin rdv Origine'] = df_rdv.loc[RDV[:-1],'Date fin'].values
InterRDV['Date début rdv Destination'] = df_rdv.loc[RDV[1:],'Date début'].values

for index, row in InterRDV.iterrows():
    InterRDV.loc[index,'Temps trajet initial'] = DuM[row['Origine Id'],row['Destination Id']]
    InterRDV.loc[index,'Distance trajet initial (km)'] = int(DiM[row['Origine Id'],row['Destination Id']])
    InterRDV.loc[index,'Temps dispo interRDV'] = temps_dispo(row)

InterRDV['Temps trajet initial'] = pd.to_timedelta(InterRDV['Temps trajet initial'],unit='m')
InterRDV['Temps trajet supplémentaire'] = pd.to_timedelta(0,unit='h')
InterRDV['Visites'] =''
InterRDV['Prospects visités'] = 0
InterRDV['Clients visités'] = 0
InterRDV['Temps restant'] = InterRDV['Temps dispo interRDV'] \
                        - InterRDV['Temps trajet initial']\
                        - InterRDV['Temps trajet supplémentaire']\
                        - datetime.timedelta(hours=2) * InterRDV['Clients visités'] \
                        - datetime.timedelta(minutes=30) * InterRDV['Prospects visités']
InterRDV
```

Voici notre dataframe initialisé, il va nous permettre d'ajouter des points de passages sur chacun de ses trajets dans la mesure du possible.

```python
lieux_potentiels_restant = np.array(sel.loc[sel["Ajouté"]==False,].index) # + len(df_rdv)
categorie_lieux_potentiels_restant = np.array((sel['Type'] =='Prospect' )* 1)

trajets = []
lieux_ajoutes = []
index = 2
for index, row in InterRDV.iterrows():
    arg_org  = InterRDV.loc[index,"Origine Id"]
    arg_dest = InterRDV.loc[index,"Destination Id"]
    trajet = [arg_org, arg_dest]   #dans la Duration matrix
    temps_restant = InterRDV.loc[index,'Temps restant']
    stop  = temps_restant < datetime.timedelta(hours=1) or len(lieux_potentiels_restant) == 0
    while not stop :
        devmin = np.inf
        argmin = None
        etape = None
        t_surplacemin = None
        imin = 0
        existe_candidat = False
        for i in range(len(lieux_potentiels_restant)):
            arg = lieux_potentiels_restant[i] + len(df_rdv)
            for t in range(len(trajet)-1) :
                dev = DuM[trajet[t],arg] + DuM[arg,trajet[t+1]] - DuM[trajet[t],trajet[t+1]]
                t_surplace = datetime.timedelta(hours=0.5) if categorie_lieux_potentiels_restant[i] else datetime.timedelta(hours=2)
                possible = InterRDV.loc[index,'Temps restant'] >= datetime.timedelta(hours=dev / 60) +  t_surplace
                if possible and dev < devmin :
                    existe_candidat = True
                    argmin = lieux_potentiels_restant[i] + len(df_rdv)    #argmin dans la Duration Matrix
                    imin = i
                    etape = t
                    devmin = dev
                    categorie = 'Client' if categorie_lieux_potentiels_restant[i] == 0 else 'Prospect'
                    t_surplacemin = t_surplace
        if existe_candidat :
            lieux_ajoutes.append(imin)
            sel.loc[imin,'Ajouté'] = True
            trajet.insert(etape+1,argmin) 
            InterRDV.loc[index,'Temps trajet supplémentaire'] += datetime.timedelta(hours=devmin / 60)
            InterRDV.loc[index,categorie+'s visités'] += 1
            InterRDV.loc[index,'Temps restant'] -= datetime.timedelta(hours=devmin / 60) + t_surplacemin
            lieux_potentiels_restant = np.delete(lieux_potentiels_restant,imin)
            categorie_lieux_potentiels_restant = np.delete(categorie_lieux_potentiels_restant,imin)
            #temps_restant = row['Temps restant']
        stop  = InterRDV.loc[index,'Temps restant'] < datetime.timedelta(hours=1) or len(lieux_potentiels_restant) == 0 or not existe_candidat
    if index == 0:
        trajets = trajet
    if index > 0:
        trajets = trajets + trajet[1:]
    InterRDV.loc[index,'Visites'] = str(trajet[1:-1])[1:-1]
```
