```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 900

X = np.random.uniform(0,1,(2,N))
plt.scatter(X[0,:],X[1,:],marker='+')
m = np.zeros((N,N))
for i in range(N):
    for j in range(i+1,N):
        m[i,j] = np.linalg.norm(X[:,i]-X[:,j])
        m[j,i] = m[i,j]
m = m/m.max()        
```

```python
plt.figure(figsize=(12,6));

ax1 = plt.subplot(121)
ys = np.sort(np.unique(m.flatten()))
xs = np.arange(len(ys))/len(ys)
ax1.plot(xs,ys)

ax2 = plt.subplot(122)
ax2.hist(m.flatten(),bins=200);
```

```python
from scipy.interpolate import interp1d
```

```python
f1 = interp1d(xs, ys, kind='nearest')
```

```python
plt.figure(figsize=(6,6));
plt.plot(xs,ys,'b')
plt.plot(xs,f1(xs),'r')
```
