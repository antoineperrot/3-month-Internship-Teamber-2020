### Import des librairies, du fichier groupe, et définition de la période de planification des tâches multi-utilisateurs :

```python
from Planning import *
group_file = 'exemple_groupe_de_travail.xlsx'
TMU_file   = 'exemple_TMF.xlsx'

DATE_DEBUT = '2020-05-11 08:30:00'
DATE_FIN   = '2020-05-16 18:00:00'
```

## On importe le groupe dont on va planifier les tâches multi-utilisateurs :

```python
group = pd.read_excel(group_file)
group
```

## On créé des objets de type planning pour travailler avec :

```python
plannings = list()
for i in range(len(group)):
    P = Planning(group['Nom'][i],
                group['ID'][i], group['Fichier Horaires'][i])
    P.initialise(DATE_DEBUT,DATE_FIN)
    P.addImperatifs(pd.read_excel(group['Fichier Impératifs'][i]))
    plannings.append(P)
```

## Import du fichier contenant les tâches multi-utilisateurs :

```python
TMU = pd.read_excel(TMU_file)
TMU.head()
```

## Planification des tâches multi-utilisateurs :

```python
planned_TMU = planifieTMU(TMU, DATE_DEBUT, DATE_FIN, plannings, maxiter = 1500)
for i  in np.unique(planned_TMU['id utilisateur'].values):
    plannings[i].addImperatifs(planned_TMU.loc[planned_TMU['id utilisateur']==i,
                                               ['Objet','Priorité','Date début',"Date fin"]])
planned_TMU
```

## Les tâches sont maintenant dans les plannings des utilisateurs, exemple :

```python
plannings[3].makePlanning(to_file=False)
```
